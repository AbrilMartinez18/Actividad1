```python
import pandas as pd
from pandas import DataFrame as df 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
barcelonaDF=pd.read_csv("listingsB.csv")
```

```python
barcelonaDF.columns
```

```python
barcelonaDF=barcelonaDF[['host_acceptance_rate','review_scores_rating', 'price', 'review_scores_cleanliness', 'availability_365','review_scores_communication','number_of_reviews','room_type']]
```

```python
barcelonaDF.head(10)
```

```python
barcelonaDF['room_type'].unique()
```

**Ajustar formato**

```python
barcelonaDF["host_acceptance_rate"]=barcelonaDF["host_acceptance_rate"].str.replace("%","")
barcelonaDF["price"] = barcelonaDF["price"].str.replace("$","")
barcelonaDF["price"] = barcelonaDF["price"].str.replace(",","")
```

```python
barcelonaDF.head(10)
```

```python
barcelonaDF.dtypes
```

**cambiar tipos de variables**

```python
barcelonaDF['host_acceptance_rate']=barcelonaDF['host_acceptance_rate'].astype({'host_acceptance_rate': 'float64'})
barcelonaDF['price']=barcelonaDF['price'].astype({'price': 'float64'})
```

```python
barcelonaDF.dtypes
```

```python
barcelonaDF.shape
```

**Dividir dataframes por tipo de habitaci√≥n**

```python
#1.PRIVATE ROOM
PR=barcelonaDF[barcelonaDF['room_type']=='Private room']

#2.ENTIRE HOME/APT
FP=barcelonaDF[barcelonaDF['room_type']=='Entire home/apt']

#3 SHARED ROOM
SR=barcelonaDF[barcelonaDF['room_type']=='Shared room']
```

##**Tipo 1. PrivateRoom**

```python
PR=PR[['host_acceptance_rate','review_scores_rating', 'price', 'review_scores_cleanliness', 'availability_365','review_scores_communication','number_of_reviews']]
```

```python
PR.head(10)
```

**Contar nulos**

```python
valores_nulos=PR.isnull().sum()
print(valores_nulos)
```

**Reemplazar valores**

```python
#remplazar valores
from numpy import nan as NaN
PR=PR.fillna({'host_acceptance_rate':float(PR['host_acceptance_rate'].mean())})
PR=PR.fillna({'review_scores_rating':float(PR['review_scores_rating'].mean())})
PR=PR.fillna({'review_scores_cleanliness':float(PR['review_scores_cleanliness'].mean())})
PR=PR.fillna({'review_scores_communication':float(PR['review_scores_communication'].mean())})
valores_nulos=PR.isnull().sum()
print(valores_nulos)
```

**OUTLIERS**

```python
y=PR

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
PR=PR[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
```

```python
valores_nulos=PR.isnull().sum()
print(valores_nulos)
```

**Llenar con la media**

```python
PR=PR.fillna({'host_acceptance_rate':float(PR['host_acceptance_rate'].mean())})
PR=PR.fillna({'review_scores_rating':float(PR['review_scores_rating'].mean())})
PR=PR.fillna({'price':float(PR['price'].mean())})
PR=PR.fillna({'review_scores_cleanliness':float(PR['review_scores_cleanliness'].mean())})
PR=PR.fillna({'review_scores_communication':float(PR['review_scores_communication'].mean())})
PR=PR.fillna({'number_of_reviews':(PR['number_of_reviews'].mean())})
```

```python
valores_nulos=PR.isnull().sum()
print(valores_nulos)
```

**correlaciones**

```python
PR.corr()
```

```python
column_sels=['host_acceptance_rate','review_scores_rating', 'price', 'review_scores_cleanliness', 'availability_365','review_scores_communication']
x=PR.loc[:,column_sels]#variables explicativas

y=PR['number_of_reviews']#target, variable dependiente

fig, axs=plt.subplots(ncols=3,nrows=2,figsize=(16,8))

axs=axs.flatten()
for i, k in enumerate(column_sels):
  sns.regplot(x=x[k],y=y,data=PR,ax=axs[i])
  plt.tight_layout(pad=0.4)
```

```python
corr=PR.corrwith(PR['number_of_reviews'])
corr
```

```python
#COEFICIENTE DE DETERMINACI√ìN R^2
coef_deter=pow(corr,2)
coef_deter
```

**Selecci√≥n de mejor modelo de Regresi√≥n Lineal Mult. considerando el coeficiente de correlaci√≥n**

---

Habitaci√≥n privada



```python
#Declaramos las variables dependientes e independientes para la regresi√≥n lineal
Vars_Indep=PR[['host_acceptance_rate','review_scores_cleanliness','review_scores_rating']]
Var_Dep= PR['number_of_reviews']
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
type(model)
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#obtener coeficientes para el modelo ajustado
model.__dict__
```

Ecuaci√≥n

---
Y=0.193357(host_acceptance_rate)+1.1299(review_scores_cleanliness)+2.0809(review_scores_rating)-23.257


**Crear el modelo matem√°tico que describa de mejor manera el n√∫mero de rese√±as**

```python
y_pred= model.predict(X=PR[['host_acceptance_rate','review_scores_cleanliness','review_scores_rating']])
y_pred
```

```python
#Coeficiente de Determinaci√≥n del modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
